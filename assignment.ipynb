```python
with open("webscrapings.csv","w",encoding="utf8",newline="") as f:
  thewriter=writer(f)
  header=["Product url","Product name","Description","ASIN","Product Description","manufacturer","ratings","reviews","price"]
  thewriter.writerow(header)
  for u in url:
    r=requests.get(u,headers=Headers)
    soup=BeautifulSoup(r.content,"html.parser")
    l=soup.find_all("div",attrs={"class":"a-section a-spacing-small a-spacing-top-small"})
    for o in l:
      arr=[]
  
      k=o.find_all("div",attrs={"class":"a-section a-spacing-none puis-padding-right-small s-title-instructions-style"})
      if k!=[]:
        for m in k:
          c=m.find_all("h2",attrs={"class":"a-size-mini a-spacing-none a-color-base s-line-clamp-2"})
          for d in c:
            e=d.find_all("a",attrs={"class":"a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"})
      
            g=("https://www.amazon.in"+e[0].get("href"))
            arr.append(g)
            for h in e:
              w=h.find("span",attrs={"class":"a-size-medium a-color-base a-text-normal"}).text
              if w=="":
                arr.append(" ")
              arr.append(w)
            e1=requests.get(g,headers=Headers)
            so=BeautifulSoup(e1.content,"html.parser")
            y=so.find_all("div",attrs={"id":"detailBullets_feature_div"})
            for hj in y:
              gh=hj.find_all("ul",attrs={"class":"a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list"})
          #print(gh)
              for yu in gh:
                kl=yu.find_all("li")
                cc={}
                for dg in kl:
                  sd=dg.find_all("span",attrs={"class":"a-list-item"})
                  he=[]
                  value=[]
                  a=[]
                  for pp in sd:
                    ff=pp.find_all("span")
                    if ff!=[]:
                      BB=ff[0].text.replace(":","").replace("\n","").replace("\u200f","").replace("\u200e","").strip()
                      he.append(BB)
                      if he[0]=='Product Dimensions':
                        bb=ff[1].text
                        if bb=="":
                          arr.append(" ")
                        arr.append(bb)
                      if he[0] == "ASIN":
                        bb=ff[1].text
                        if bb=="":
                          arr.append(" ")
                        arr.append(bb)
                      if he[0]=="Item model number":
                        bb=ff[1].text
                        if bb=="":
                          arr.append(" ")
                        arr.append(bb)  
                      else:
                        pass
            a=""
            for hj in y:
              gh=hj.find_all("ul",attrs={"class":"a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list"})
          #print(gh)
              for yu in gh:
                kl=yu.find_all("li")
                cc={}
                for dg in kl:
                  sd=dg.find_all("span",attrs={"class":"a-list-item"})
                  he=[]
                  value=[]
                  for pp in sd:
                    ff=pp.find_all("span")
                    if ff!=[]:
                      BB=ff[0].text.replace(":","").replace("\n","").replace("\u200f","").replace("\u200e","").strip()
                      he.append(BB)
                      if he[0]=='Manufacturer':
                        bb=ff[1].text
                        a+=bb
            arr.append(a)
            
      p=o.find_all("div",attrs={"class":"a-section a-spacing-none a-spacing-top-micro"})
      for j in p:
        h=j.find_all("div",attrs={"class":"a-row a-size-small"})
        for k in h:
          q=k.find_all("span")
          m=q[0].find("span",attrs={"class":"a-size-base"})
          c=(m.text)
          if c=="":
            arr.append(" ")
          arr.append(c)
          d=k.find_all("span")
          for i in d:
            x=i.find_all("a",attrs={"class":"a-link-normal s-underline-text s-underline-link-text s-link-style"})
            if x!=[]:
              for t in x:
                v=t.find("span").text
                i=(v.replace("(","").replace(")",""))
                if i=="":
                  arr.append(" ")
                arr.append(i)
      n=o.find_all("div",attrs={"class":"sg-row"})
      for j in n:
        m=j.find_all("div",attrs={"class":"sg-col sg-col-4-of-12 sg-col-4-of-16 sg-col-4-of-20 sg-col-4-of-24"})
        for k in m:
          d=k.find_all("div",attrs={"class":"sg-col-inner"})
          for g in d:
            e=g.find_all("div",attrs={"class":"a-section a-spacing-none a-spacing-top-micro s-price-instructions-style"})
            for f in e:
              s=f.find_all("div",attrs={"class":"a-row a-size-base a-color-base"})
              for t in s:
                r=t.find_all("span",attrs={"class":"a-price"})
                price=r[0].find("span",attrs={"class":"a-offscreen"}).text
                if price=="":
                  arr.append(" ")
                arr.append(price)
      thewriter.writerow(arr)        
```
